# Introduction to Plotting with matplotlib.pyplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(io='../Data/plotting_data.xls', sheet_name='anscombe', index_col='ID')
df.head(3)
```

```python
# plot only the 'I' data set
fltr = df['dataset'] == 'I' # we now have a new DataFrame
fltr
df_I = df[fltr] # we now have a new DataFrame containing only 'I' data
df_I
xI = df_I['x']
yI = df_I['y']
# plot
plt.plot(xI, yI, 'x') # or 'o' etc
```

```python
# plot data for 'II' using a line chart 
# also 'III' and 'IV' as separate plots
fltr2 = df['dataset'] == 'II' # we now have a new DataFrame
fltr2
df_II = df[fltr2] # we now have a new DataFrame containing only 'I' data
df_II
xII = df_II['x']
yII = df_II['y']
plt.plot(xII, yII, 'x')

```

```python
# also 'III' and 'IV' as separate plots
fltr3 = df['dataset'] == 'III' # we now have a new DataFrame
fltr3
df_III = df[fltr3] # we now have a new DataFrame containing only 'I' data
df_III
xIII = df_III['x']
yIII = df_III['y']
plt.plot(xIII, yIII, 'x')
```

```python
# also 'III' and 'IV' as separate plots
fltr4 = df['dataset'] == 'IV' # we now have a new DataFrame
fltr4
df_IV = df[fltr4] # we now have a new DataFrame containing only 'I' data
df_IV
xIV = df_IV['x']
yIV = df_IV['y']
plt.plot(xIV, yIV, 'x')
```
